```python
#  Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#  Load the Dataset
df = pd.read_csv("customers_data.csv")
```

```python
#  View Basic Information
print("Dataset Info:")
print(df.info())

#  Conclusion:
# Dataset has no missing columns and expected datatypes.
# Gender is categorical, rest are numeric.
```

```python
# View First Few Records
print("\nFirst 5 Rows:")
print(df.head())

# Conclusion:
# Data appears clean and follows the expected schema.
```

```python
# Check for Missing Values
print("Missing Values:")
print(df.isnull().sum())

# Conclusion:
# There are no missing values in any column.
```

```python
# Summary Statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Conclusion:
# - Age ranges are reasonable (e.g., 18–70).
# - Spending Score ranges from 1 to 100.
# - Income shows a wide range (low to high earners).
```

```python
# Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=20, kde=True, color='teal')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.tight_layout()
plt.show()

# Conclusion:
# Most customers are in the age range of 20–40.
```

```python
# Annual Income Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['Annual_Income'], bins=20, kde=True, color='orange')
plt.title("Annual Income Distribution")
plt.xlabel("Income (k$)")
plt.tight_layout()
plt.show()

# Conclusion:
# Income is fairly normally distributed, with clusters around mid-income levels.
```

```python
# Spending Score Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['Spending_Score'], bins=20, kde=True, color='purple')
plt.title("Spending Score Distribution")
plt.xlabel("Score")
plt.tight_layout()
plt.show()

# Conclusion:
# Spending Score is evenly distributed, indicating a variety of spending behaviors.
```

```python
# Pairplot for Relationships
sns.pairplot(df[['Age', 'Annual_Income', 'Spending_Score']])
plt.suptitle("Pairwise Relationships", y=1.02)
plt.tight_layout()
plt.show()

# Conclusion:
# Some clusters or trends may be visible — helpful for future clustering models.
```

```python
# Age vs Spending Score
plt.figure(figsize=(6, 4))
sns.boxplot(x='Age', y='Spending_Score', data=df)
plt.title("Spending Score by Age")
plt.tight_layout()
plt.show()

# Conclusion:
# This can reveal gender-based differences in spending behavior (if any).
```

```python
# Income vs Spending Score
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Annual_Income', y='Spending_Score', hue='Age', data=df)
plt.title("Income vs Spending Score")
plt.tight_layout()
plt.show()

# Conclusion:
# Certain customer segments spend more despite having lower income.
# Great input for clustering (e.g., k-means).
```
