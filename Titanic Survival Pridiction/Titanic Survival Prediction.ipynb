```python
# Perform EDA with step-by-step comments and conclusions included as comments
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df=pd.read_csv('titanic_data.csv')
```

```python
# 1. Dataset Shape
df.shape  # (891, 12)
# Conclusion:
# Dataset contains 891 rows and 12 columns.
```

```python
# 2. Dataset Info
df.info()
# Conclusion:
# 'Age', 'Cabin', and 'Embarked' columns contain missing values.
```

```python
# 3. First 5 Rows
df.head()
# Conclusion:
# The dataset includes features like Name, Sex, Age, Pclass, etc., typical for Titanic passenger data.
```

```python
# 4. Missing Values
df.isnull().sum()
#  Conclusion:
# - Cabin has 687 missing values (~77%), may be dropped.
# - Age has 177 missing values (~19%), can be imputed.
# - Embarked has only 2 missing values.
```

```python
# 5. Summary Statistics
df.describe()
#  Conclusion:
# - Fare is highly skewed, with a maximum of 512.
# - Age varies widely, from 0.4 to 80.
# - SibSp and Parch mostly range from 0 to 3.
```

```python
# 6. Class Distribution (Target Variable)
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
# Conclusion:
# About 38% survived, and 62% did not—dataset is imbalanced.
```

```python
# 7. Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
# Conclusion:
# Females had significantly higher survival rates compared to males.
```

```python
# 8. Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
# Conclusion:
# 1st class passengers had the highest survival rates; 3rd class the lowest.
```

```python
# 9. Age Distribution by Survival
sns.histplot(df['Age'],bins=20,kde=True)
plt.title('Age Distribution by Survival')
plt.show()
# Conclusion:
# passengers between age 20-35 were more likely to survive.
```

```python
# 11. Feature Engineering - Family Size
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
sns.histplot(df['FamilySize'],bins=20,kde=True)
plt.show()
# Conclusion:
# Very large families had lower survival rates; small to medium families fared better.
```
