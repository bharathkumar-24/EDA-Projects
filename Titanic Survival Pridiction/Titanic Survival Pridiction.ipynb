```python
# titanic_data_analysis.py
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("train.csv")
```

```python
# Basic Info
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# Data Cleaning
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Feature Engineering
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
# Visualization 1: Survival Count
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
# Visualization 2: Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# Visualization 3: Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# Visualization 4: Age Distribution by Survival
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Visualization 5: Survival by Embarked Port
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Port of Embarkation')
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# Visualization 6: Survival vs Family Size
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.show()
```
